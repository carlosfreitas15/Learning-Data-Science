```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z= x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y) 

#o argumento alpha é para a transparencia da linha
#TODO ESTE CONJUNTO TEM DE FICAR JUNTO
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y)
ax2.plot(x,y)

#o argumento alpha é para a transparencia da linha
#TODO ESTE CONJUNTO TEM DE FICAR JUNTO
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x,z)
ax2.plot(x,y)
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

#o argumento alpha é para a transparencia da linha
#TODO ESTE CONJUNTO TEM DE FICAR JUNTO
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(4,3))

axes[0].plot(x,y, lw=3,ls='--')

axes[1].plot(x,z, color="red")

plt.tight_layout()
```
