```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #mostrar dentro do jupiter
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='black',linewidth=2,ls='-',marker='o',markersize=7,
       markerfacecolor='red',markeredgewidth=1,markeredgecolor='black') 

#o argumento alpha Ã© para a transparencia da linha
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='black',linewidth=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
