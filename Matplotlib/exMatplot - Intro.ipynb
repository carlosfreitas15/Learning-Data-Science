```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
#plt.shot() #podia colocar esta linha e seria = a fazer um print
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OBJECT ORIENTED
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #criamos um canva sem nada

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
```

```python
fig = plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8]) #dimensoes do plot
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) #todos estes valores tm de ser entre 0 e 1 

#o primeiro valor da lista  a distancia da esquerda
#o segundo  distancia de baixo
#o terceiro e o quarto  a largura e altura do plot

axes1.plot(x,y)
axes1.set_title('BIGGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2) #nos subplots podemos dizer o n de linhas e colunas
#cria 1 linha e duas colunas de plots. Se for 3, 3 criamos 9 plots

#for current_ax in axes:
    #current_ax.plot(x,y)
    
axes[0].plot(x,y)
axes[1].plot(y,x)
```
