```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #mostrar dentro do jupiter
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure(figsize=(8,2)) #o primeiro argumento de figsize é largura


ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(8,2)) #o primeiro argumento de figsize é largura



axes.plot(x,y)
```

```python
#a linha 12 e a 14 são iguais
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
#ao estabeler o nº de linhas e colunas, criamos um array/matriz de graficos
#sendo que podemos aceder aos mesmos pelo seu index
```

```python
fig.savefig('my_picture.png',dpi=100) #dpi aumenta detalhe 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cute')

ax.legend(loc=0) #isto so funciona se adicionarmos o argumento de label em cada
#invocação do plot

#o argumento loc=0 identifica a melhor posição para a legenda
#ver documentação
#tambem podemos colocar as posições (como fizemos com o subplot)
```
