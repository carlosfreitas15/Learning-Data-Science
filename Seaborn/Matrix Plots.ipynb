```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
#para criar heatmaps precisamos de converter estas tabelas em matrizes
#porque a coluna do index nao tem designação/variavel
```

```python
tc = tips.corr() #correlacionamos os dados
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=3)
```

```python
sns.clustermap(fp,cmap='coolwarm',linecolor='white',linewidths=3)
```

```python
#o cluster, coloca os anos mais semelhantes entre si (consoante os dados)
```

```python
sns.clustermap(fp,cmap='coolwarm',linecolor='white',linewidths=3,standard_scale=1)
```

```python
#com o standardscale, conseguimos quase dividir o ano em epocas
#Jan, Fev e Nov sao os meses com menos viagens, e Jul Aug com mais
```
