```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
#compara o total gasto por genero
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#compara o desvio padrao do total da conta por genero
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#os pontinhos fora dos quartis, sao outliers
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
#compara 3 variaveis: total bill com o dia e se  fumador ou n
```

```python
 sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#mais dificil de visualizar, principalmente para quem no  familiar
```

```python
 sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```
